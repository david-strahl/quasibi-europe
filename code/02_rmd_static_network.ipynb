# Static RMD Correlation Calculation

This script will calculate the static RMD correlation coefficients for the downsampled data.

```python
# load all the time series
ys = np.load("../data/data_downsampled_binned.npy")
```

```python
%%time

# calculate pairwise pearson correlation coefficients
rmd = pairwise_distances(ys, metric=quasibi.rmd)
```

```python
# save the pearson matrix
quasibi.save_symmat(rmd, "../output/rmd_correlation_matrix")
```

```python
rmd_abs = np.abs(rmd)
plt.imshow(rmd_abs > np.quantile(rmd_abs, 0.995), cmap="binary")
```
