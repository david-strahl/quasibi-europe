```python
import numpy as np
import matplotlib.pyplot as plt
import quasibi
```

```python
t = np.linspace(0, 2*np.pi, 100)
x = np.sin(t) + 0.1*np.random.random(100)
y = np.sin(t) + 0.1*np.random.random(100)

plt.plot(t, x)
plt.plot(t, y)
plt.show()

quasibi.pearson(x, y)
```

```python
quasibi.pearson(x, 0.5*y)
```

```python
quasibi.pearson(0.5*x, 0.5*y)
```

```python
quasibi.pearson(x, y+0.2)
```

```python
quasibi.pearson(x+0.2, y+0.2)
```

```python
quasibi.pearson(x, 0.5*y + 0.2*t)
```

```python
plt.plot(t, x + 0.5*t)
plt.plot(t, y + 0.5*t)

quasibi.pearson(x + 0.5*t, y + 0.5*t)
```
